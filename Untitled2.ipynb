```python
import pandas as pd
import torch
from transformers import BertTokenizer, BertForSequenceClassification
from transformers import pipeline
from bs4 import BeautifulSoup
import requests
```

```python
# Load scraped text data
data = pd.read_csv('scraped_data.csv')  # Replace with your file path
texts = data['text']  # Assuming the scraped text is in the 'text' column
```

```python
# Load pre-trained BERT model and tokenizer for binary classification
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
model = BertForSequenceClassification.from_pretrained("bert-base-uncased", num_labels=2)

# Use pipeline for text classification
classifier = pipeline("text-classification", model=model, tokenizer=tokenizer)
```

```python
def classify_text(text):
    result = classifier(text)
    label = result[0]['label']  # 'LABEL_0' for no dark pattern, 'LABEL_1' for dark pattern
    confidence = result[0]['score']
    return label, confidence

# Apply the classifier to each scraped text
data['label'], data['confidence'] = zip(*data['text'].apply(classify_text))
```

```python
# Save the results to a new CSV file
data.to_csv('classification_results.csv', index=False)

# Display the results
print(data)
```

```python
def scrape_website(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    # Extract all text content
    text = soup.get_text(separator=" ")
    return text

# Example usage
url = 'https://example-ecommerce.com'
scraped_text = scrape_website(url)

# Classify the scraped text
label, confidence = classify_text(scraped_text)
print(f"Label: {label}, Confidence: {confidence}")
```
